```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('../feature_extraction/Word2Vec.csv',index_col='docno')
dataset
```

Enter doc id or query point with required number of neighbours

```python
Docid = ' LA122989-0001 '
neighbours=50
distance_frame=dataset
```

```python
query_point=distance_frame[distance_frame.index==Docid]
query_point
```

```python
from scipy.spatial import distance
ary = distance.cdist(distance_frame, query_point, metric='euclidean')
distance_frame['dist']=ary
distance_frame
```

```python
sorted_distance=distance_frame.sort_values('dist',ascending=True)
sorted_distance
```
