```python
from sklearn.manifold import TSNE
from numpy import reshape
import seaborn as sns
import pandas as pd  
```

```python
tsne = TSNE(n_components=2, verbose=1, random_state=123)
principalComponents = tsne.fit_transform(df)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
print(principalDf)
```

```python
y = pd.DataFrame(y)
y.columns=['class']
finalDf = pd.concat([principalDf,y], axis = 1)
finalDf

sns.scatterplot(x="principal component 1", y="principal component 2", hue=finalDf['class'].tolist(),
                palette=sns.color_palette("hls", 3),
                data=finalDf).set(title="data T-SNE projection")
```

```python
import pandas as pd
df=pd.read_csv("tf_idf_features+sentiment.csv")
df=pd.DataFrame(df, columns=df.columns)
y = df.loc[:,['class']].values
df=df.drop(columns='docno')
df=df.drop(columns='class')
df
```

```python
from sklearn.manifold import TSNE
from numpy import reshape
import seaborn as sns
import pandas as pd 

tsne = TSNE(n_components=2, verbose=1, random_state=123)
principalComponents = tsne.fit_transform(df)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
print(principalDf)
```

```python

y = pd.DataFrame(y)
y.columns=['class']
finalDf = pd.concat([principalDf,y], axis = 1)
finalDf

sns.scatterplot(x="principal component 1", y="principal component 2", hue=finalDf['class'].tolist(),
                palette=sns.color_palette("hls", 3),
                data=finalDf).set(title="data T-SNE projection")
```
